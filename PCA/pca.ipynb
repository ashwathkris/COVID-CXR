```python
features_df.loc[345]
features_df.index.values
copy_df = features_df.drop(index = features_df.index[345], axis = 0)
y = y.drop(index = y.index[345], axis=0)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(copy_df, y, test_size=0.2, random_state=0)
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
pca = PCA(n_components = 40)
X_train = pca.fit_transform(X_train)
X_test = pca.transform(X_test)
```

```python
explained_variance = pca.explained_variance_ratio_
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(26,16))
plt.scatter(x=[i+1 for i in range(len(explained_variance))],
            y=explained_variance,
           s=200, alpha=0.75,c='orange',edgecolor='k')
plt.grid(True)
plt.title("Explained variance ratio of the \nfitted principal component vector\n",fontsize=25)
plt.xlabel("Principal components",fontsize=15)
plt.xticks([i+1 for i in range(len(explained_variance))],fontsize=15)
plt.yticks(fontsize=15)
plt.ylabel("Explained variance ratio",fontsize=20)
plt.show()
```
